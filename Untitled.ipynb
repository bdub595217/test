```python
from sklearn.datasets import fetch_mldata
mnist = fetch_mldata('MNIST original')
```

```python
import numpy as np

np.random.seed(42)

m = 10000
idx = np.random.permutation(60000)[:m]
```

```python
X = mnist['data'][idx]
y = mnist['target'][idx]
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=42)
```

```python
mnist_tsne = tsne.fit_transform(X)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,10))
plt.scatter(mnist_tsne[:,0], mnist_tsne[:,1], c=y, cmap="jet", marker=".")
plt.colorbar()
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2, random_state=42)
```

```python
mnist_pca = pca.fit_transform(X)
```

```python
plt.figure(figsize=(12,10))
plt.scatter(mnist_pca[:,0], mnist_pca[:,1], c=y, cmap="jet", alpha=.5, marker=".")
plt.colorbar()
plt.show()
```

```python
import matplotlib
```

```python
plt.figure(figsize=(9,9))
cmap = matplotlib.cm.get_cmap("jet")
for digit in (2, 3, 5):
    plt.scatter(mnist_tsne[y == digit, 0], mnist_tsne[y == digit, 1], c=cmap(digit / 9))
plt.axis('off')
plt.show()
```

```python
idx = (y == 2) | (y == 3) | (y == 5) 
X_subset = X[idx]
y_subset = y[idx]

tsne_subset = TSNE(n_components=2, random_state=42)
X_subset_reduced = tsne_subset.fit_transform(X_subset)
```

```python
plt.figure(figsize=(9,9))
for digit in (2, 3, 5):
    plt.scatter(X_subset_reduced[y_subset == digit, 0], X_subset_reduced[y_subset == digit, 1], c=cmap(digit / 9))
plt.axis('off')
plt.show()
```
